## Particle Swarm
In this script we run particle swarm over the estabilished parameters of the OMC using our NN as a stand in for traditional FDTD solvers.

```python
import numpy as np
from pyswarm import pso
import tensorflow as tf
```

```python
model_path = r"Z:\lwoodard\Courses\Deep Learning\BDTNet\optcial_Q_Net\conv_model_20241201-204952.keras"
model = tf.keras.models.load_model(model_path)
rand_params = np.zeros((11, 18, 3))
rand_params[:, :, 2] = 1
rand_params = rand_params[np.newaxis, :]
prediction = model.predict(rand_params, verbose = 0)
prediction[0,0]
```

```python
# Define the objective function to maximize
def objective_function(X):
    X = X.reshape(11, 18, 3)
    rand_params = X[np.newaxis, :]
    prediction = model.predict(rand_params, verbose = 0)
    out = prediction[0,0]
    # return -out because pyswarm minimizes functions
    return -out


def gen_rand_input(std):
    rand_params = np.zeros((11, 18, 3))
    rand_params[:, :, 2] = 1
    rand_params += np.random.normal(0, std, size = (11, 18, 3))
    return rand_params

# give this a whirl to see the prediction for different raandom OMCs
inp = gen_rand_input(0.001)
inp = inp.flatten()
objective_function(inp)
```

```python
# Define bounds for the input variables (X)
rand_params = np.zeros((11, 18, 3))
rand_params[:, :, 2] = 1
nominal = rand_params

lb = (nominal - 0.01).flatten()
ub = (nominal + 0.01).flatten()

# Run PSO using the pso function from pyswarm
best_position, best_value = pso(objective_function, lb, ub)

# Since pyswarm minimizes the objective function, we need to negate the result
print("Best Position: ", best_position)
print("Best Value: ", -best_value)  # Negate to get the actual maximized value
```
